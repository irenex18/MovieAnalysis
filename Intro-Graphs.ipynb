```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
industry = pd.read_csv("Desktop/EastCoastDataOpen/movie_industry.csv", encoding = "latin-1")
movies = pd.read_csv("Desktop/EastCoastDataOpen/movies.csv")
ratings = pd.read_csv("Desktop/EastCoastDataOpen/ratings.csv")

movies.dropna(inplace=True)
ratings.dropna(inplace=True)
industry.dropna(inplace=True)
```

```python
movies['year'] = movies.title.str.extract("\((\d{4})\)", expand=True)
uniqueGenres = pd.DataFrame(movies.genres.str.split('|').tolist()).stack().unique()
ratings.loc[:, 'time']=pd.to_datetime(ratings['timestamp'], unit='s')
ratings.loc[:, 'year'] = ratings.loc[:,'time'].dt.year
```

```python
ratings['year']
```

```python
moviesReleased = movies[['movieId','year']].groupby('year')
plt.plot(moviesReleased.year.unique(), moviesReleased.movieId.nunique(), "r-o")
plt.title('Movies per Year')
plt.ylabel('Number of Movies Released')
plt.xlabel('Year')
```

```python
ratingsWritten = ratings[['rating', 'year']].groupby('year')
plt.plot(ratingsWritten.year.unique(), ratingsWritten.count(), "g-o")
plt.title('Ratings per Year')
plt.ylabel('Number of Ratings')
plt.xlabel('Year')
```
